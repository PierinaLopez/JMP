# Exploratory Data Analysis
This notebook visualizes the Titanic dataset to:
1. Understand the distribution of numerical features (e.g., age, fare).
2. Identify correlations between features.
3. Spot any outliers that may need handling.

```python
# Import libraries for visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
# This is the dataset after basic cleaning
df = pd.read_csv('titanic_cleaned.csv')

# Visualize the distribution of 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Visualize the distribution of 'Fare'
plt.figure(figsize=(8, 6))
sns.histplot(df['fare'], bins=20, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# Display a correlation heatmap to identify relationships between features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
