# AWS SageMaker Model
This notebook uploads the prepared Titanic dataset to AWS S3 and trains an XGBoost model using AWS SageMaker.

```python
# Import libraries
import boto3
import sagemaker
from sagemaker import get_execution_role

# Initialize SageMaker session
# This session manages interactions with AWS SageMaker
sagemaker_session = sagemaker.Session()
role = get_execution_role()
print("SageMaker session initialized successfully.")
# Upload dataset to S3 for use in SageMaker
bucket_name = 'titanic-pierina-lopez'
s3 = boto3.client('s3')
file_name = 'titanic_prepared.csv'
s3.upload_file(file_name, bucket_name, file_name)

print(f"Uploaded {file_name} to S3 bucket {bucket_name}.")

# Define XGBoost Estimator
# This specifies the model and training parameters for SageMaker
xgboost = sagemaker.estimator.Estimator(
    image_uri=sagemaker.image_uris.retrieve('xgboost', sagemaker_session.boto_region_name),
    role=role,
    instance_count=1,
    instance_type='ml.m5.large',
    output_path=f's3://{bucket_name}/output',
    sagemaker_session=sagemaker_session
)

# Train the model using the dataset in S3
xgboost.fit({'train': f's3://{bucket_name}/{file_name}'})
```
